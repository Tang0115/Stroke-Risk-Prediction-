```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

METRICS_DIR  = os.path.join('..', '..', 'results', 'metrics')
FIGURES_PATH = os.path.join('..', '..', 'results', 'figures')
```

## 1. Aggregate All Phase Metrics

```python
p2 = pd.read_csv(os.path.join(METRICS_DIR, 'phase2_baseline_metrics.csv'))
p3 = pd.read_csv(os.path.join(METRICS_DIR, 'phase3_model_comparison.csv'))

all_metrics = pd.concat([p2, p3], ignore_index=True)
display(all_metrics.sort_values('recall_stroke', ascending=False))
```

## 2. Comparison Bar Chart

```python
plot_df = all_metrics.set_index('model')[['recall_stroke', 'f1_stroke', 'roc_auc']]
plot_df.plot(kind='bar', figsize=(12, 5))
plt.title('Model Comparison: Recall, F1, and ROC-AUC (stroke class)')
plt.ylabel('Score')
plt.xticks(rotation=30, ha='right')
plt.ylim(0, 1)
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_PATH, 'p4_model_comparison.png'), dpi=150)
plt.show()
```
